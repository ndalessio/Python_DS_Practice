#### Variación del peso de los pinguinos a lo largo de los años

```python
pinguinos.groupby("anio").masa_corporal_g.agg("mean")
```

```python
pinguinos["mc_anios"] = pinguinos["masa_corporal_g"].groupby(pinguinos['anio']).transform("mean")
pinguinos["anio"] = pd.to_datetime(pinguinos["anio"], format = "%Y")
```

```python
pinguinos.head()
```

```python
ax = sns.relplot(
        data = pinguinos, 
        x = "anio", y = "mc_anios")

ax.set(title = "Variación anual del peso de los pinguinos")
ax.set(xlabel = "Año", ylabel = "Peso en gramos")
```

```python
pinguinos["mc_anios_especie"] = pinguinos.groupby(['anio', 'especie'])['masa_corporal_g'].transform("mean")
pinguinos.head()
```

```python
pinguinos.mc_anios_especie.value_counts()
```

#### Variación anual del peso de los pinguinos por especie

```python
ax = sns.relplot(
        data = pinguinos, 
        x = "anio", y = "mc_anios_especie", hue="especie")


ax.set(title = "Variación anual del peso de los pinguinos por especie")
ax.set(xlabel = "Año", ylabel = "Peso en gramos")

#tengo que ver cómo achicar el rango de los ticks en el eje x

#ax.set(xticks= pinguinos.anio)
#ax.set(xlim=(1, 5))
#ax.tick_params(axis = 'x', which = 'major', length = 5)
#FacetGrid.set(xticks=np.arange(1,4,1))
```
